```python
# Method 1: with a temporary variable
x = 10
y = 20
print("Before swap:", x, y)

# use a temporary holder
temp = x
x = y
y = temp
print("After swap (with temp):", x, y)

# Method 2: without a temporary variable
x = 10  
y = 20
print("\nBefore swap:", x, y)
x, y = y, x       
print("After swap (no temp):", x, y)
```

```python
# right angled triangle
for i in range(1, 6):
    print("*" * i)
```

```python
#inverted right angled triangle
for i in range(5, 0, -1):
    print("*" * i)
```

```python
# numerical pattern
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# reverse numerical pattern
for i in range(5, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```

```python
# repeated numbers pattern
for i in range(1, 6):
    print(str(i) * i)
```

```python
# pyramid pattern
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2*i - 1))
```

```python
# inverted pyramid pattern
n = 5
for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * (2*i - 1))
```

```python
#square pattern
n = 5
for i in range(n):
    print("* " * n)
```

```python
# alphabet pattern
for i in range(1, 6):
    for j in range(i):
        print(chr(65 + j), end=" ")
    print()
```

```python
# combination pattern
for i in range(1, 5):
    for j in range(i):
        print(i, end=" ")
    for k in range(i):
        print("*", end=" ")
    print()
    
    
```

```python
n = 4

for i in range(1, n + 1):
    print('*' * i)
    print(str(i) * i)
```

```python
num = 1

for i in range(1, 6):
    for j in range(1, i + 1):
        if (i + j) % 2 == 0:
            print("*", end=" ")
        else:
            print(num, end=" ")
            num += 1
        
    print()
```
